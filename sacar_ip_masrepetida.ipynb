 ## importar librerias
 
    

```python
from collections import Counter


# importamos la libreria counter para sacar coincidencias de una lista
```

## cargamos el archivo con la informacion  o data log


```python
dataLog = open("failed.txt", "r")   ## cargo el archivo como si fuera una lista
```

```python
def calcularIp(autLog): # defino una funcion que recibe como parametro un string si cumple con las condiciones me
    #devuelve una IP
    primerPocisionIp = (autLog.find('from') + 5)    # localizo en la primer linea donde empieza a contar la Ip

    segundaPocisionIP = (autLog.find('port') - 1)
    ip = (autLog[primerPocisionIp :segundaPocisionIP])
    return ip
```

```python
lineas = dataLog.readlines() # leo cada uno de las lineas del archivo y las meto en una lista  es una lista de string


```

```python
print(type(lineas))
```

```python
len(lineas)
```

## Tenemos 7946 registros  o intentos de autentificacion 

```python
dataLog.close()
```

```python
listaDeIp = map(calcularIp, lineas) #a cada una de las lineas del archivo que hay en la vble linea se
# le aplica la funcion  calcular ip
```

```python
print(type(listaDeIp))
```

```python
# print(list(listaDeIp))
```

## tenemos las IP del archivo, ahora debemos buscar las de mayor coincidencia 


```python
c = Counter(listaDeIp)  # con la libreria Counter sacamos esto facilmente 
```

```python
print (c)
```
